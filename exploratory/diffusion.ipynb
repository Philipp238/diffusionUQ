```python
from scipy.stats import kstest, uniform
from scipy import stats
# Perform KS test
x = stats.norm.rvs(size=100, loc=0)
ks_stat, p_value = kstest(x, stats.uniform.cdf)
print(f"KS Statistic: {ks_stat}, p-value: {p_value}")
```
